```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.model_selection import train_test_split
import pathlib
from random import sample
import torch
import pandas as pd
import seaborn as sns
```

```python
transform = A.Compose([
    A.Resize(256, 256, p = 1.),
])
```

```python
source_image_tensors = []
source_mask_tensors = []
has_stroke_masks = []
has_stroke_images = []
```

```python

dataset_path = './data/brainMRI_copy.h5'

with h5py.File(dataset_path, 'r') as dataset:
    images = dataset['images'][:]
    masks = dataset['masks'][:]
```

```python
for idx, image in enumerate(images):
    image = image.astype(np.float32)
    mask = masks[idx].squeeze().astype(np.float32)

    image = image.transpose(1, 2, 0)
    augmented = transform(image=image, mask=mask)
    image = augmented['image']
    mask = augmented['mask']
    source_image_tensors.append(image.transpose(2, 0, 1))
    source_mask_tensors.append(mask)

    
```

```python
source_image_tensors_train, source_image_tensors_test, source_mask_tensors_train, source_mask_tensors_test = train_test_split(source_image_tensors, source_mask_tensors, test_size=0.2, random_state = 42)
```

```python
def countplot_class_elements(without_stroke, with_stroke):
    data = {
        'Class': ['Нет инсульта', 'Есть инсульт',],
        'Count': [without_stroke, with_stroke]
    }
    
    # Create a DataFrame
    df = pd.DataFrame(data)
    
    # Create a bar plot
    plt.figure(figsize=(8, 6))

    def make_autopct(values):
        def my_autopct(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
        return my_autopct
    
    plt.pie(df['Count'], labels=data['Class'], autopct=make_autopct(data['Count'])) 
```
